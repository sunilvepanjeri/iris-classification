```python
from sklearn.datasets import load_iris
idata=load_iris()
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(idata.data,idata.target,test_size=0.2,random_state=12)
```

```python
from lazypredict.Supervised import LazyClassifier
```

```python
cl=LazyClassifier(verbose=0,ignore_warnings=True,custom_metric=None)
model,predictions=cl.fit(xtrain,xtest,ytrain,ytest)
```
